# Part A--Translation

## R code 

x <- c(1, 3, 5, 7)

y <- ifelse(x > 4, x^2, x + 1)

mean(y)

## Python equivalent 

```python
import numpy as np
x = np.array([1, 3, 5, 7])
y = np.where(x > 4, x**2, x + 1)
print(np.mean(y))
```

# Part B 

## List vs Numpy array

```python
#creating a python list 
list1 = [1,2,3,4,5]
print(f"Python list multiplication result: {list1*2}")

#converting the list to a numpy array
list2= np.array(list1)
print(f"Numpy array multiplication result:{list2*2}")
```

- Each operation return a different result. Multiplying the python list by 2 just duplicate the element of the list, making it two times longer. However, when the same operation is performed on the numpy list, it only doubles each element.

-  Multiplying a python list by an integer simply concatenates copies of the list. NumPy arrays perform fast element‑wise numerical multiplication instead of structural repetition. Because lists are general containers and NumPy arrays are mathematical objects. 


# Part C--Reflection 

When you run the R code, it creates a numeric vector, and y <- ifelse(x > 4, x^2, x + 1) applies the condition element‑wise, producing a new vector. A similar Python version would use NumPy to create and perform the same  conditional transformation. 

The structure is nearly parallel, but R’s ifelse is a built-in function, while Python relies on NumPy for the vectorized conditional logic.
