## **Non-parametric inference**


**Methodology:**

To determine the 'typical' price of a laptop without being misled by extreme luxury models, we used a Bootstrapping method. We took a 10% random sample (130 laptops) and simulated 10,000 different market scenarios to calculate a reliable median price.

**Results of the Inference:**

Sample Median Price: 50,189.76

95% Confidence Interval: [43,023.60, 57,096.45]

**Interpretation:**

 Because the laptop market contains highly expensive model that inflate the average price, the median gives us a much more accurate picture of what a typical customer spends. We can state with 95% of confidence that the true middle-point of the entire laptop market falls between roughly 43,000 and 57,000.

## **Summary and recommendation**

This project analyzed laptop sales data, revealing that Dell dominates sales volume, while high-end brands like Razer and LG command average prices, contrasting with budget-friendly options from Acer and Lenovo. A strong positive correlation exists between RAM capacity with laptop prices, whereas physical size (inches) shows no significant relationship. Through bootstrapping, the median laptop price was estimated to be between 43,000 and 57,000 Indian Rupee, offering a more accurate market representation than the average price which is often skewed by luxury models.

For future analysis, it would be beneficial to delve deeper into the specific features or market strategies that differentiate the high-end brands, as well as investigate the factors contributing to the low sales volumes of brands like Huawei and Google. Additionally, building a predictive model to forecast laptop prices based on the identified key specifications (RAM, memory) could provide valuable insights for pricing and product development.
