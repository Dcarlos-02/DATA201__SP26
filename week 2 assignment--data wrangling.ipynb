## PART A

```python
#creating summary table 

summary = (df.query('price>250000 and size>1000').assign(price_per_sqft=lambda x : x['price']/x['size']).groupby('neighborhood')['price_per_sqft'].agg(avg_price_per_sqft='mean',median_price_per_sqft='median', count='count').sort_values('avg_price_per_sqft', ascending=False))
summary
```

## PART B--dplyr equivalent

#R code equivalent

summary = df |> 
           filter(df, df.price>250000, df.size>1000) |>
           
           mutate(price_per_sqft=df.price/df.size) |>
           
           groupby(neighborhood) |>
           
           summarise(
           avg_price_per_sqft = mean(price_per_sqft),
           median_price_per_sqft = median(price_per_sqft),
           count = n()
           ) |>
           
           arrange(desc(avg_price_per_sqft))


           N.B replace the dots(.) by the dollar sign('$'). I code it with the dots because when I run it with the dollar sign, it disappears.

## PART D

1. Why must we wrap each condition in parentheses when using `&` in pandas?
2. What is the advantage of method chaining over creating many temporary DataFrames?
3. In `.agg(mean_price=("price", "mean"))`, what does `"price"` represent? What does `"mean"` represent?
4. When you `groupby("neighborhood")`, why does `neighborhood` appear on the left (index) in the result table?

### ANSWERS

1. We must wrap the conditions because of operator precedence. Some operators bind tighter than others. Without parentheses, python evaluates certain operation first before getting to the next. Parentheses force the resolution of conditions and then combine them.
2. Using chaining allows one to avoid declaring variables that they will not reuse. In addition, it is less likely that a variable is used in place of another.
3. In this code, price represent the input variable to be aggregated, and mean represents the aggregation function.
4. Because neighborhood is used to organize the data. It becomes the identifier of each group. Each row represents a group.
